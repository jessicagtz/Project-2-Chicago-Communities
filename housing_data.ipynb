```python
import pandas as pd
import numpy as np
```

```python
df_housing = pd.read_csv('Final Tables through 2016 - Housing.csv').fillna(0)
```

```python
df_housing
```

```python
areas = []
for i in range(0, len(df_housing)):
    if df_housing.iloc[i]['Home Ownership in Chicago Community Areas'] == 0:
        areas.append(df_housing.iloc[i+1]['Home Ownership in Chicago Community Areas'])
        
areas
```

```python
thirty = []
fourty = []
fifty = []
sixty = []
seventy = []
eighty = []
ninety = []
thousand = []
ten = []

for i in range(0, len(df_housing)):
    if df_housing.iloc[i]['Home Ownership in Chicago Community Areas'] == 'Percent owner occupied':
        thirty.append(df_housing.iloc[i][1])
        fourty.append(df_housing.iloc[i][2])
        fifty.append(df_housing.iloc[i][3])
        sixty.append(df_housing.iloc[i][4])
        seventy.append(df_housing.iloc[i][5])
        eighty.append(df_housing.iloc[i][6])
        ninety.append(df_housing.iloc[i][7])
        thousand.append(df_housing.iloc[i][8])
        ten.append(df_housing.iloc[i][9])
```

```python
df_housingclean = pd.DataFrame({
    'Community':areas,
    '1930':thirty,
    '1940':fourty,
    '1950':fifty,
    '1960':sixty,
    '1970':seventy,
    '1980':eighty,
    '1990':ninety,
    '2000':thousand,
    '2010':ten
})
```

```python
df_housingclean
```

```python
community_id = df_housingclean.Community.str.split('.', expand = True)
community_id.head()
```

```python
housing_clean = pd.DataFrame({
    'ID':community_id[0],
    '1930':thirty,
    '1940':fourty,
    '1950':fifty,
    '1960':sixty,
    '1970':seventy,
    '1980':eighty,
    '1990':ninety,
    '2000':thousand,
    '2010':ten
})
housing_clean.head()
```

```python
housing_clean = housing_clean.set_index('ID')
housing_clean
```
