```python
import pandas as pd
import numpy as np
```

```python
asian = pd.read_csv('asian_pop_id.csv')
asian = asian[["ID","2010"]]
asian.columns = ["ID","2010 Asian Pop"]
# asian
```

```python
white = pd.read_csv('white_pop_id.csv')
white = white[["ID","2010"]]
white.columns = ["ID","2010 White Pop"]
# white
```

```python
other = pd.read_csv('other_pop_id.csv')
other = other[["ID","2010"]]
other.columns = ["ID","2010 Other Pop"]
# other
```

```python
black = pd.read_csv('black_pop_id.csv')
black = black[["ID","2010"]]
black.columns = ["ID","2010 Black Pop"]
# black
```

```python
hispanic = pd.read_csv('hispanic_pop_id.csv')
hispanic = hispanic[["ID","2010"]]
hispanic.columns = ["ID","2010 Hispanic Pop"]
# hispanic
```

```python
race = asian.merge(white,on='ID').merge(other,on='ID').merge(black,on='ID').merge(hispanic,on='ID')
```

```python
race.to_csv('race_2010_master.csv',index=False)
```
