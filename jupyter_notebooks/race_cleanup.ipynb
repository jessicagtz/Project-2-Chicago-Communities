```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Final Tables through 2016 - Race & Ethnicity.csv').fillna(0)
```

```python
df
```

```python
areas = []
for i in range(0, len(df)):
    if df.iloc[i]['Race and Ethnicity in Chicago Community Areas'] == 0:
        areas.append(df.iloc[i+1]['Race and Ethnicity in Chicago Community Areas'])
        
areas
```

```python
thirty = []
fourty = []
fifty = []
sixty = []
seventy = []
eighty = []
ninety = []
thousand = []
ten = []

for i in range(0, len(df)):
    if df.iloc[i]['Race and Ethnicity in Chicago Community Areas'] == 'Other':
        thirty.append(df.iloc[i][1])
        fourty.append(df.iloc[i][2])
        fifty.append(df.iloc[i][3])
        sixty.append(df.iloc[i][4])
        seventy.append(df.iloc[i][5])
        eighty.append(df.iloc[i][7])
        ninety.append(df.iloc[i][8])
        thousand.append(df.iloc[i][9])
        ten.append(df.iloc[i][10])
        
# ten
# thirty
```

```python
df_clean = pd.DataFrame({
    'Community':areas,
    '1930':thirty,
    '1940':fourty,
    '1950':fifty,
    '1960':sixty,
    '1970':seventy,
    '1980':eighty,
    '1990':ninety,
    '2000':thousand,
    '2010':ten
})
```

```python
# df_clean = df_clean.set_index('Community')
df_clean
```

```python
community_id = df_clean.Community.str.split('.', expand = True)
community_id.head()
```

```python
clean = pd.DataFrame({
    'ID':community_id[0],
    '1930':thirty,
    '1940':fourty,
    '1950':fifty,
    '1960':sixty,
    '1970':seventy,
    '1980':eighty,
    '1990':ninety,
    '2000':thousand,
    '2010':ten
})
clean.head()
```

```python
clean = clean.set_index('ID')
clean
```

```python
clean = clean[:-1]
clean
```

```python
clean.to_csv('other_pop_id.csv')
```
